```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
import time
import random
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
def keyboard_nav(tab_pressed_number, time_spanX, time_span_Y):
    # Cr√©er une instance de la classe ActionChains
    actions = ActionChains(driver)
    
    # Effectuer 5 fois l'appui sur la touche Tab avec des d√©lais al√©atoires
    for _ in range(tab_pressed_number):
        actions.send_keys(Keys.TAB)
        actions.pause(random.uniform(time_spanX, time_spanX))  # D√©lai al√©atoire entre 0.5 et 0.8 secondes
    actions.send_keys(Keys.RETURN)
    # Ex√©cuter les actions
    actions.perform()
```

```python
def scroll_down(max_scroll):
    iteration = 0
    while iteration < max_scroll:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
        time.sleep(random.uniform(2.5, 3.1))
        iteration += 1
```

```python
def get_videos_data():
    
    videos = []
    
    try:
        # Utilisez BeautifulSoup pour analyser le contenu de la page une fois qu'elle a √©t√© enti√®rement charg√©e
        soup = bs(driver.page_source, 'html.parser')
        videos_elements = soup.select("div[data-e2e='explore-item-list'] > div")
    
        for video in videos_elements:
            try:
                tags = [tag.text for tag in video.select("div[data-e2e='explore-card-video-caption'] a")]
            except:
                tags = []
    
            try:
                author = video.select("[data-e2e='explore-card-user-unique-id']")[0].text
            except:
                author = ""
    
            try : 
                like = video.select("div[data-e2e='explore-card-like-container'] strong")[0].text
            except:
                like = 0
            
            try : 
                link = video.select("div[data-e2e='explore-item'] > div > div > a")[0].get("href")
            except:
                link = ""
                
            try : 
                description = video.select("div[data-e2e='explore-item'] > div > div > a > div div:nth-child(2) div div:last-child")[0].text
            except:
                description = ""

            try : 
                caption = video.select("div[data-e2e='explore-card-video-caption'] > div")[0].text
            except:
                caption = ""
    
            try : 
                views = video.select("div[data-e2e='explore-item'] > div > div > a > div div:nth-child(3) strong")[0].text
            except:
                try:
                    views = video.select("div[data-e2e='explore-item'] > div > div > a > div div:nth-child(2) strong")[0].text
                except:
                    views = ""
            element = {
                "tags" : tags,
                "author" : author,
                "like" : like,
                "description" : description,
                "link" : link,
                "view" : views, 
                "caption" : caption
            }
            videos.append(element)
            
        return videos

            
    except Exception as e:
        print(f"Une exception s'est produite dans la r√©cup√©ration des vid√©os: {str(e)}")        
```

```python
url = "https://www.tiktok.com/explore"

# Initialisation du navigateur
driver = webdriver.Chrome()  # Assurez-vous d'avoir le driver appropri√© pour votre navigateur
driver.get(url)
try: 
    # Continue as guest by tab navigation
    keyboard_nav(8, 0.5, 0.8)

    # 17 = sport / 15 = Chant et danse
    keyboard_nav(17, 0.8, 1.3)

except Exception as e:
    print(f"Une exception s'est produite dans la navigation: {str(e)}")
    
try: 
    scroll_down(20)
    
except Exception as e:
    print(f"Une exception s'est produite dans le scroll: {str(e)}")

```
